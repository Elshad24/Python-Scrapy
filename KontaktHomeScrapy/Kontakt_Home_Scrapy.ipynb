```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import warnings
warnings.filterwarnings('ignore')
# pd.set_option('display.max_rows', 5000)
# pd.set_option('display.max_columns', 5000)
# pd.set_option('display.width', -1)
```

```python
url = 'https://kontakt.az/'
content = requests.get(url)
soup = BeautifulSoup(content.text,'html.parser')
```

```python
url_1 = []
a = soup.find_all('div',class_='menu-open menu-open-item')
for i in a:
    for j in i.find_all('a',class_='menuliparent_a'):
        if j.get('href') in url_1 or len(j.get('href').split('/')) > 6:
            break
        else:
            url_1.append(j.get('href'))
```

```python
lst = []
df = pd.DataFrame()
for url in url_1:
    print(url.split('/')[4])
    content_1 = requests.get(url)
    soup_2 = BeautifulSoup(content_1.text,'html.parser')
    try:
        b = soup_2.find('span',class_='pages').text.split(' / ')[1]
        a = soup_2.find('span',class_='pages').text.split(' / ')[0]
    except:
        b = 2
        a = 1
    for i in range(int(a),int(b)+1):
        contents = requests.get(url + 'page/' + str(i) + '/')
        soup_1 = BeautifulSoup(contents.text,'html.parser')
        category_name = url.split('/')[3]
        sub_category = url.split('/')[4]
        for i in soup_1.find_all('div',class_='cart-body-top'):
            try:
                price = float(i.find('div', class_='cart-footer').find('span').text.replace(' M',''))
            except:
                price = None
            try:
                discount = i.find('div', class_='pleft').text.split('-')[1].replace('M','')
                discount_type = i.find('div', class_='pleft').text.split('-')[0]
            except:
                discount_type = None 
                discount = None

            dc={'Category_Name':category_name,
                'Sub_Category':sub_category,
                'Name':i.find('div', class_ = 'name').find('a').text.strip(),
                'Price':price,
                'Discount_Type': discount_type,
                'Discount': discount
                }
            lst.append(dc)
df = df.append(lst)
df.to_excel('Kontakt_Home_Scrapy.xlsx')
```
